## Deiken Sklklar ve Sat Fiyatlar likisi

Kategorilerin sat fiyatlar ile ilikilerini incelemek adna bir fonksiyon yazdktan sonra, her kesikli deikenin skln ve sat fiyatlarn ortalamasnn grselletirmek iin de bir fonksiyon yazacaz.

```python
#Kategori oranlar ve kategori bana den ev fiyatlar ortalamalar
def calculate_mean_target_per_category(df, var):

    # kategori bana den ev says
    temp_df = pd.Series(df[var].value_counts() / len(df)).reset_index()
    temp_df.columns = [var, 'ev_oran']

    # kategori bana den ev fiyatlar ortalamas
    temp_df = temp_df.merge(df.groupby([var])['SalePrice'].mean().reset_index(),
                            on=var,
                            how='left')

    return temp_df
```

```python
#Fonksiyon rnek
calculate_mean_target_per_category(house, 'Garage Cars')
```

```python
#Yukardaki bilgileri grselletiren fonksiyon
def plot_categories(df, var):
    
    fig, ax = plt.subplots(figsize=(8, 4))
    plt.xticks(df.index, df[var], rotation=90)

    ax2 = ax.twinx()
    ax.bar(df.index, df['ev_oran'], color='lightgrey')
    ax2.plot(df.index, df['SalePrice'], color='green', label='Seconds')
    ax.axhline(y=0.05, color='red')
    ax.set_ylabel('Kategori bana ev oran')
    ax.set_xlabel(var)
    ax2.set_ylabel('Kategori bana den ev fiyat ortalamalar')
    plt.show()
```

```python
#Fonksiyon rnek
plot_categories(calculate_mean_target_per_category(house, 'Garage Cars'), 'Garage Cars')
```

```python
#Tm kesikli deikenlerin sklk ve sat fiyat ilikisi grselletirmeleri
for col in kesikliler:        
    temp_df = calculate_mean_target_per_category(house, col)
    plot_categories(temp_df, col)
```

ou deikenin kategori sklklar bakmndan olduka az sayda olduklarn, dolaysyla train-test ayrmnda belli kategorilerin sadece train ya da test setinde yer alabilecei ihtimaline karn **"ender"** kategori birletirme ilemi uygulamamz gerekmektedir. Burada ender kategori birleimi iin baz deikenlerin sklnn yetmemesi sebebiyle **0.05 oran eik deerini 0.1'e** ykseltmek daha salkl olacaktr.

# 3) Veri n leme Pipelinelar

Bu aamaya kadar hangi deikenle ilgili ne tr bir veri n ileme aksiyonu alacamz belirledik. Bu n ilemelerin hepsi birer **"hiperparametre"** olarak kabul edilebilir. Yani buraya kadar not aldmz ve uygulayacamz tekniklerin hepsinin yerine baka teknikler de kullanlabilir.

**Dipnot:** Veri n ileme, deikenlerin mmkn olduunca bir yapay renme algoritmas tarafndan dzgn ekilde kullanlabilmesi iin yaplr. Aklayc deikenler ile baml deikenler arasnda anlaml ilikiler zaten yoksa, veri n ilemeden dramatik performans artlar iin bir **mucize bekleyemeyeceimizi** syleyebiliriz.

```python
X_train, X_test, y_train, y_test = train_test_split(house.drop('SalePrice', axis = 1), house['SalePrice'], 
                                                    test_size = 0.2, random_state = 42)

X_train.shape, X_test.shape
```

Burada kullanacamz deiken saylar bizi u alardan endielendirmemesi gerekiyor:

1. **Alan bilgisi:** Amerika'da emlak deerlemeleri yaplrken evin tm zelliklerinin fiyata etki edebilecei bilgisi elimizde mevcut.

2. **Pipeline'da karlacak deikenler:** Pipeline esnasnda modele katks olamayacan dndmz ve teknik olarak bilgi tamayan (dominant veri, ou eksik veri) verileri kartacaz.

3. **Elastic Net:** Hiperparametrelerini kontrol edeceimiz iin, hem l1 hem de l2 dzeltmesiyle katsaylar cezalandran bu algoritma, hem oklu balanm (multicollinearity) problemi stesinden gelebilir. Hem de aklayc olmayan deikenlerin etkilerini 0'a indirerek kendi iinde deiken seimi de yapabilir.

4. **Random Forest:** Hiperparametrelerini kontrol edeceimiz iin, overfittingi nlemek amacyla, algoritmadaki aalarn ok dallanmasna izin vermeyeceiz. Belli bir yere kadar byyebilecek aalar, algoritmann doas gerei kendi iinde nemli deikenleri seerek belirli derinlie kadar byyecektir.

5. **KNN ve SVM:** Hiperparametrelerini kontrol edeceimiz iin ve lineer-olmayan algoritmalar olduklar iin, herhangi bir varsaym olmadan bu algoritmalar kullanabiliriz.


## 3.1) Lineer Pipeline

Pipelanlar olutururken listelerin **+ (art)** sembol ile birletirilebilme zellii sayesinde kolayca n ilemelerimizi yapabileceiz.

```python
#Liste birletirme rnek
surekli_cikar + kesikli_cikar + kategorik_cikar
```

```python
#Lineer modellere uygun ve kullanlabilecek n ilemeleri yapalm
lineer_pipeline = Pipeline([
    
    #==karlacak deikenler==
    ('cikan_degiskenler', sel.DropFeatures(
        features_to_drop = surekli_cikar + kesikli_cikar + kategorik_cikar
    )),
    
    #==Eksik verilerin doldurulmas==
    
    #Eksik verilerin iaretlenmesi
    ('missing_isaret', imp.AddMissingIndicator(
        variables = surekli_missing_ind
    )),
    
    #Medyan doldurma
    ('medyan_doldurma', imp.MeanMedianImputer(
        imputation_method = 'median', variables = surekli_medyan_doldurma
    )),
    
    
    #Kategorik "missing" ile doldurma
    ('missing_doldurma', imp.CategoricalImputer(
        imputation_method='missing', variables=kategorik_missing_doldurma, return_object=True
    )),
    
    #Mod ile doldurma
    ('frequent', imp.CategoricalImputer(
        imputation_method='frequent', variables=kategorik_mod, return_object=True
    )),
    
    #==Aykr deerler==
    #Aralkta tutma
    ('aykiri_deger', out.Winsorizer(
        capping_method = 'iqr', variables = surekli_out                
    )),
    
    #==Kesikletirme==
    ('kesiklestirme', disc.EqualFrequencyDiscretiser(
        variables = surekli_kesik, return_object=True
    )),
    
    #==Ender kategori gruplama==
    #Kesikliler iin 0.1
    ('ender_kesikli', enc.RareLabelEncoder(
        tol = 0.1, n_categories = 1, variables = kesikli_rare
    )),
    
    #Dierleri iin 0.05
    ('ender_diger', enc.RareLabelEncoder(
        tol = 0.05, n_categories = 0, variables = kategorik_rare + surekli_kesik
    )),
    
    
    #==Kategorik veri kodlama==
    ('encoder', enc.OneHotEncoder(
    drop_last = True, variables = surekli_kesik + kesikli_rare + kategorik_rare
    )),
    
    
    #==Deiken dnm==
    ('donusum', tf.YeoJohnsonTransformer(variables=surekli_out)),
    
    #==Deiken lekleme==
    ('olcekleme', StandardScaler())
])
```

```python
#Pipeline fit edip modelleme iin hazr olan veri setimizi inceleyelim
lineer_pipeline.fit(X_train, y_train)

#Transfrom metodu ile train ve test veri setlerimizii renilen n ileme parametrelerine gre dntrelim
X_train_lineer = lineer_pipeline.transform(X_train)
X_test_lineer = lineer_pipeline.transform(X_test)
X_test_lineer
```

## 3.2) Monoton Pipeline

```python
#Monoton ilikilerden faydalanabilecek modeller iin n ilemeleri yapalm
monoton_pipeline = Pipeline([
    
    #==karlacak deikenler==
    ('cikan_degiskenler', sel.DropFeatures(
        features_to_drop = surekli_cikar + kesikli_cikar + kategorik_cikar
    )),
    
    #==Eksik verilerin doldurulmas==
    
    #Eksik verilerin iaretlenmesi
    ('missing_isaret', imp.AddMissingIndicator(
        variables = surekli_missing_ind
    )),

    #Medyan doldurma
    ('medyan_doldurma', imp.MeanMedianImputer(
        imputation_method = 'median', variables = surekli_medyan_doldurma
    )),
    
    
    #Kategorik "missing" ile doldurma
    ('missing_doldurma', imp.CategoricalImputer(
        imputation_method='missing', variables=kategorik_missing_doldurma, return_object=True
    )),
    
    #Mod ile doldurma
    ('frequent', imp.CategoricalImputer(
        imputation_method='frequent', variables=kategorik_mod, return_object=True
    )),
    
    #==Kesikletirme==
    ('kesiklestirme', disc.EqualFrequencyDiscretiser(
        variables = surekli_out + surekli_kesik, return_object=True
    )),
    
    #==Ender kategori gruplama==
    #Kesikliler iin 0.1
    ('ender_kesikli', enc.RareLabelEncoder(
        tol = 0.1, n_categories = 1, variables = kesikli_rare
    )),
    
    #Dierleri iin 0.05
    ('ender_diger', enc.RareLabelEncoder(
        tol = 0.05, n_categories = 0, variables = kategorik_rare + surekli_kesik + surekli_out
    )),
    
    #==Kategorik veri kodlama==
    ('encoder', enc.OrdinalEncoder(
        encoding_method='ordered', variables = surekli_out + surekli_kesik + kesikli_rare + kategorik_rare
    )),
        
    #==Deiken lekleme==
    ('olcekleme', StandardScaler())

])
```

```python
#Pipeline fit edip modelleme iin hazr olan veri setimizi inceleyelim
monoton_pipeline.fit(X_train, y_train)

#Transfrom metodu ile train ve test veri setlerimizii renilen n ileme parametrelerine gre dntrelim
X_train_monoton = monoton_pipeline.transform(X_train)
X_test_monoton = monoton_pipeline.transform(X_test)
X_test_monoton
```

## 3.3) Aa Pipeline

```python
#Karar aac karar noktalarndan faydalanabilecek modeller iin n ilemeleri yapalm
agac_pipeline = Pipeline([
    
    #==karlacak deikenler==
    ('cikan_degiskenler', sel.DropFeatures(
        features_to_drop = surekli_cikar + kesikli_cikar + kategorik_cikar
    )),
    
    #==Eksik verilerin doldurulmas==
    
    #Eksik verilerin iaretlenmesi
    ('missing_isaret', imp.AddMissingIndicator(
        variables = surekli_missing_ind
    )),
    
    #Medyan doldurma
    ('medyan_doldurma', imp.MeanMedianImputer(
        imputation_method = 'median', variables = surekli_medyan_doldurma
    )),
    
    
    #Kategorik "missing" ile doldurma
    ('missing_doldurma', imp.CategoricalImputer(
        imputation_method='missing', variables=kategorik_missing_doldurma, return_object=True
    )),
    
    #Mod ile doldurma
    ('frequent', imp.CategoricalImputer(
        imputation_method='frequent', variables=kategorik_mod, return_object=True
    )),
    
    
    #==Ender kategori gruplama==
    #Kesikliler iin 0.1
    ('ender_kesikli', enc.RareLabelEncoder(
        tol = 0.1, n_categories = 1, variables = kesikli_rare
    )),
    
    #Dierleri iin 0.05
    ('ender_diger', enc.RareLabelEncoder(
        tol = 0.05, n_categories = 0, variables = kategorik_rare
    )),
    
    #==Kategorik veri kodlama==
    #Aa kodlama
    ('encoder', enc.DecisionTreeEncoder(
        cv=3, scoring='neg_mean_squared_error',
        regression=True, random_state=42, variables=kesikli_rare + kategorik_rare
    )),
    
    #==Kesikletirme==
    #Aa kesikletirme
    ('kesiklestirme', disc.DecisionTreeDiscretiser(
        cv=3, scoring='neg_mean_squared_error', variables=surekli_out + surekli_kesik,
        regression=True, random_state=42
    )),
    
    #==Deiken lekleme==
    ('olcekleme', StandardScaler())
    
])
```

```python
#Pipeline fit edip modelleme iin hazr olan veri setimizi inceleyelim
agac_pipeline.fit(X_train, y_train)

#Transfrom metodu ile train ve test veri setlerimizii renilen n ileme parametrelerine gre dntrelim
X_train_agac = agac_pipeline.transform(X_train)
X_test_agac = agac_pipeline.transform(X_test)
X_test_agac
```

Farkl n ilemelerden geen veri setlerimiz, makine renmesi modellemeleri iin hazr!

## Modellerin Kyaslanmas

Modelleri cross-validation sonularna gre kyaslamak iin, ilgili objeden en iyi ortalama mse ve standart sapmay eken fonksiyonu yazalm.

```python
#En iyi skorlar eken fonksiyon
def select_best_score(grid):
    
    results = pd.DataFrame(grid.cv_results_)
    
    results.sort_values(by='mean_test_score', ascending=False, inplace=True)
    
    results = results[['mean_test_score', 'std_test_score']]
    
    return results.head(1)

#Fonksiyonu test edelim
select_best_score(rf_lineer)
```
