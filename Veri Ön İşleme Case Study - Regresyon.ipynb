## Değişken Sıklıkları ve Satış Fiyatları İlişkisi

Kategorilerin satış fiyatları ile ilişkilerini incelemek adına bir fonksiyon yazdıktan sonra, her kesikli değişkenin sıklığını ve satış fiyatlarını ortalamasının görselleştirmek için de bir fonksiyon yazacağız.

```python
#Kategori oranları ve kategori başına düşen ev fiyatları ortalamaları
def calculate_mean_target_per_category(df, var):

    # kategori başına düşen ev sayısı
    temp_df = pd.Series(df[var].value_counts() / len(df)).reset_index()
    temp_df.columns = [var, 'ev_oran']

    # kategori başına düşen ev fiyatları ortalaması
    temp_df = temp_df.merge(df.groupby([var])['SalePrice'].mean().reset_index(),
                            on=var,
                            how='left')

    return temp_df
```

```python
#Fonksiyon örnek
calculate_mean_target_per_category(house, 'Garage Cars')
```

```python
#Yukarıdaki bilgileri görselleştiren fonksiyon
def plot_categories(df, var):
    
    fig, ax = plt.subplots(figsize=(8, 4))
    plt.xticks(df.index, df[var], rotation=90)

    ax2 = ax.twinx()
    ax.bar(df.index, df['ev_oran'], color='lightgrey')
    ax2.plot(df.index, df['SalePrice'], color='green', label='Seconds')
    ax.axhline(y=0.05, color='red')
    ax.set_ylabel('Kategori başına ev oranı')
    ax.set_xlabel(var)
    ax2.set_ylabel('Kategori başına düşen ev fiyatı ortalamaları')
    plt.show()
```

```python
#Fonksiyon örnek
plot_categories(calculate_mean_target_per_category(house, 'Garage Cars'), 'Garage Cars')
```

```python
#Tüm kesikli değişkenlerin sıklık ve satış fiyatı ilişkisi görselleştirmeleri
for col in kesikliler:        
    temp_df = calculate_mean_target_per_category(house, col)
    plot_categories(temp_df, col)
```

Çoğu değişkenin kategori sıklıkları bakımından oldukça az sayıda olduklarını, dolayısıyla train-test ayrımında belli kategorilerin sadece train ya da test setinde yer alabileceği ihtimaline karşın **"ender"** kategori birleştirme işlemi uygulamamız gerekmektedir. Burada ender kategori birleşimi için bazı değişkenlerin sıklığının yetmemesi sebebiyle **0.05 oran eşik değerini 0.1'e** yükseltmek daha sağlıklı olacaktır.

Ayrıca, **havuz alanı** değişkeninin neredeyse tamamamının 0 değeri alması sebebiyle satış fiyatıyla ilgili herhangi bir bilgi taşımayacağından, veri setimizden **çıkarabiliriz**.

Son olarak, evin ortalama satış fiyatlarının her yıl düşmesi, satış fiyatlarını açıklayabilecek değişkenlerden biri olabilir, fakat modele dahil etmememizin ve veri setimizden çıkartmamızın asıl sebebi, y-eksenine dikkatlica baktığımızda fiyatların çok dar bir alanda değişkenlik göstermesi diyebiliriz. Yani, yıllar arasında ortalama fiyatlar bakımından ciddi **farklar göremiyoruz.**

Aynı yorumu satış ayları için de söyleyebiliriz. Fakat, kompleks modeller kuracağımız için modelin kendi içinde bu değişken ile ilgili kararı vermesini isteyebiliriz. Dolayısıyla, ender kategori kodlaması sonrası, modellere ekleyebiliriz.

# 3) Veri Ön İşleme Pipelineları

Bu aşamaya kadar hangi değişkenle ilgili ne tür bir veri ön işleme aksiyonu alacağımızı belirledik. Bu ön işlemelerin hepsi birer **"hiperparametre"** olarak kabul edilebilir. Yani buraya kadar not aldığımız ve uygulayacağımız tekniklerin hepsinin yerine başka teknikler de kullanılabilir.

**Dipnot:** Veri ön işleme, değişkenlerin mümkün olduğunca bir yapay öğrenme algoritması tarafından düzgün şekilde kullanılabilmesi için yapılır. Açıklayıcı değişkenler ile bağımlı değişkenler arasında anlamlı ilişkiler zaten yoksa, veri ön işlemeden dramatik performans artışları için bir **mucize bekleyemeyeceğimizi** söyleyebiliriz.

## Train ve Test Setlerine Ayırma

```python
#Kesikli değişkenleri object veri tipine dönüştürme
house[kesikliler] = house[kesikliler].astype('object')
```

```python
X_train, X_test, y_train, y_test = train_test_split(house.drop('SalePrice', axis = 1), house['SalePrice'], 
                                                    test_size = 0.2, random_state = 42)

X_train.shape, X_test.shape
```

Burada kullanacağımız değişken sayıları bizi şu açılardan endişelendirmemesi gerekiyor:

1. **Alan bilgisi:** Amerika'da emlak değerlemeleri yapılırken evin tüm özelliklerinin fiyata etki edebileceği bilgisi elimizde mevcut.

2. **Pipeline'da çıkarılacak değişkenler:** Pipeline esnasında modele katkısı olamayacağını düşündüğümüz ve teknik olarak bilgi taşımayan (dominant veri, çoğu eksik veri) verileri çıkartacağız.

3. **Elastic Net:** Hiperparametrelerini kontrol edeceğimiz için, hem l1 hem de l2 düzeltmesiyle katsayıları cezalandıran bu algoritma, hem çoklu bağlanım (multicollinearity) problemi üstesinden gelebilir. Hem de açıklayıcı olmayan değişkenlerin etkilerini 0'a indirerek kendi içinde değişken seçimi de yapabilir.

4. **Random Forest:** Hiperparametrelerini kontrol edeceğimiz için, overfittingi önlemek amacıyla, algoritmadaki ağaçların çok dallanmasına izin vermeyeceğiz. Belli bir yere kadar büyüyebilecek ağaçlar, algoritmanın doğası gereği kendi içinde önemli değişkenleri seçerek belirli derinliğe kadar büyüyecektir.

5. **KNN ve SVM:** Hiperparametrelerini kontrol edeceğimiz için ve lineer-olmayan algoritmalar oldukları için, herhangi bir varsayım olmadan bu algoritmaları kullanabiliriz.


## 3.1) Lineer Pipeline

Pipelanları oluştururken listelerin **+ (artı)** sembolü ile birleştirilebilme özelliği sayesinde kolayca ön işlemelerimizi yapabileceğiz.

```python
#Liste birleştirme örnek
surekli_cikar + kesikli_cikar + kategorik_cikar
```

```python
#Lineer modellere uygun ve kullanılabilecek ön işlemeleri yapalım
lineer_pipeline = Pipeline([
    
    #==Çıkarılacak değişkenler==
    ('cikan_degiskenler', sel.DropFeatures(
        features_to_drop = surekli_cikar + kesikli_cikar + kategorik_cikar
    )),
    
    #==Eksik verilerin doldurulması==
    
    #Eksik verilerin işaretlenmesi
    ('missing_isaret', imp.AddMissingIndicator(
        variables = surekli_missing_ind
    )),
    
    #Medyan doldurma
    ('medyan_doldurma', imp.MeanMedianImputer(
        imputation_method = 'median', variables = surekli_medyan_doldurma
    )),
    
    
    #Kategorik "missing" ile doldurma
    ('missing_doldurma', imp.CategoricalImputer(
        imputation_method='missing', variables=kategorik_missing_doldurma, return_object=True
    )),
    
    #Mod ile doldurma
    ('frequent', imp.CategoricalImputer(
        imputation_method='frequent', variables=kategorik_mod, return_object=True
    )),
    
    #==Aykırı değerler==
    #Aralıkta tutma
    ('aykiri_deger', out.Winsorizer(
        capping_method = 'iqr', variables = surekli_out                
    )),
    
    #==Kesikleştirme==
    ('kesiklestirme', disc.EqualFrequencyDiscretiser(
        variables = surekli_kesik, return_object=True
    )),
    
    #==Ender kategori gruplama==
    #Kesikliler için 0.1
    ('ender_kesikli', enc.RareLabelEncoder(
        tol = 0.1, n_categories = 1, variables = kesikli_rare
    )),
    
    #Diğerleri için 0.05
    ('ender_diger', enc.RareLabelEncoder(
        tol = 0.05, n_categories = 0, variables = kategorik_rare + surekli_kesik
    )),
    
    
    #==Kategorik veri kodlama==
    ('encoder', enc.OneHotEncoder(
    drop_last = True, variables = surekli_kesik + kesikli_rare + kategorik_rare
    )),
    
    
    #==Değişken dönüşümü==
    ('donusum', tf.YeoJohnsonTransformer(variables=surekli_out)),
    
    #==Değişken ölçekleme==
    ('olcekleme', StandardScaler())
])
```

```python
#Pipeline fit edip modelleme için hazır olan veri setimizi inceleyelim
lineer_pipeline.fit(X_train, y_train)

#Transfrom metodu ile train ve test veri setlerimizii öğrenilen ön işleme parametrelerine göre dönüştürelim
X_train_lineer = lineer_pipeline.transform(X_train)
X_test_lineer = lineer_pipeline.transform(X_test)
X_test_lineer
```

## 3.2) Monoton Pipeline

```python
#Monoton ilişkilerden faydalanabilecek modeller için ön işlemeleri yapalım
monoton_pipeline = Pipeline([
    
    #==Çıkarılacak değişkenler==
    ('cikan_degiskenler', sel.DropFeatures(
        features_to_drop = surekli_cikar + kesikli_cikar + kategorik_cikar
    )),
    
    #==Eksik verilerin doldurulması==
    
    #Eksik verilerin işaretlenmesi
    ('missing_isaret', imp.AddMissingIndicator(
        variables = surekli_missing_ind
    )),

    #Medyan doldurma
    ('medyan_doldurma', imp.MeanMedianImputer(
        imputation_method = 'median', variables = surekli_medyan_doldurma
    )),
    
    
    #Kategorik "missing" ile doldurma
    ('missing_doldurma', imp.CategoricalImputer(
        imputation_method='missing', variables=kategorik_missing_doldurma, return_object=True
    )),
    
    #Mod ile doldurma
    ('frequent', imp.CategoricalImputer(
        imputation_method='frequent', variables=kategorik_mod, return_object=True
    )),
    
    #==Kesikleştirme==
    ('kesiklestirme', disc.EqualFrequencyDiscretiser(
        variables = surekli_out + surekli_kesik, return_object=True
    )),
    
    #==Ender kategori gruplama==
    #Kesikliler için 0.1
    ('ender_kesikli', enc.RareLabelEncoder(
        tol = 0.1, n_categories = 1, variables = kesikli_rare
    )),
    
    #Diğerleri için 0.05
    ('ender_diger', enc.RareLabelEncoder(
        tol = 0.05, n_categories = 0, variables = kategorik_rare + surekli_kesik + surekli_out
    )),
    
    #==Kategorik veri kodlama==
    ('encoder', enc.OrdinalEncoder(
        encoding_method='ordered', variables = surekli_out + surekli_kesik + kesikli_rare + kategorik_rare
    )),
        
    #==Değişken ölçekleme==
    ('olcekleme', StandardScaler())

])
```

```python
#Pipeline fit edip modelleme için hazır olan veri setimizi inceleyelim
monoton_pipeline.fit(X_train, y_train)

#Transfrom metodu ile train ve test veri setlerimizii öğrenilen ön işleme parametrelerine göre dönüştürelim
X_train_monoton = monoton_pipeline.transform(X_train)
X_test_monoton = monoton_pipeline.transform(X_test)
X_test_monoton
```

## 3.3) Ağaç Pipeline

```python
#Karar ağacı karar noktalarından faydalanabilecek modeller için ön işlemeleri yapalım
agac_pipeline = Pipeline([
    
    #==Çıkarılacak değişkenler==
    ('cikan_degiskenler', sel.DropFeatures(
        features_to_drop = surekli_cikar + kesikli_cikar + kategorik_cikar
    )),
    
    #==Eksik verilerin doldurulması==
    
    #Eksik verilerin işaretlenmesi
    ('missing_isaret', imp.AddMissingIndicator(
        variables = surekli_missing_ind
    )),
    
    #Medyan doldurma
    ('medyan_doldurma', imp.MeanMedianImputer(
        imputation_method = 'median', variables = surekli_medyan_doldurma
    )),
    
    
    #Kategorik "missing" ile doldurma
    ('missing_doldurma', imp.CategoricalImputer(
        imputation_method='missing', variables=kategorik_missing_doldurma, return_object=True
    )),
    
    #Mod ile doldurma
    ('frequent', imp.CategoricalImputer(
        imputation_method='frequent', variables=kategorik_mod, return_object=True
    )),
    
    
    #==Ender kategori gruplama==
    #Kesikliler için 0.1
    ('ender_kesikli', enc.RareLabelEncoder(
        tol = 0.1, n_categories = 1, variables = kesikli_rare
    )),
    
    #Diğerleri için 0.05
    ('ender_diger', enc.RareLabelEncoder(
        tol = 0.05, n_categories = 0, variables = kategorik_rare
    )),
    
    #==Kategorik veri kodlama==
    #Ağaç kodlama
    ('encoder', enc.DecisionTreeEncoder(
        cv=3, scoring='neg_mean_squared_error',
        regression=True, random_state=42, variables=kesikli_rare + kategorik_rare
    )),
    
    #==Kesikleştirme==
    #Ağaç kesikleştirme
    ('kesiklestirme', disc.DecisionTreeDiscretiser(
        cv=3, scoring='neg_mean_squared_error', variables=surekli_out + surekli_kesik,
        regression=True, random_state=42
    )),
    
    #==Değişken ölçekleme==
    ('olcekleme', StandardScaler())
    
])
```

```python
#Pipeline fit edip modelleme için hazır olan veri setimizi inceleyelim
agac_pipeline.fit(X_train, y_train)

#Transfrom metodu ile train ve test veri setlerimizii öğrenilen ön işleme parametrelerine göre dönüştürelim
X_train_agac = agac_pipeline.transform(X_train)
X_test_agac = agac_pipeline.transform(X_test)
X_test_agac
```

Farklı ön işlemelerden geçen veri setlerimiz, makine öğrenmesi modellemeleri için hazır!

## Modellerin Kıyaslanması

Modelleri cross-validation sonuçlarına göre kıyaslamak için, ilgili objeden en iyi ortalama mse ve standart sapmayı çeken fonksiyonu yazalım.

```python
#En iyi skorları çeken fonksiyon
def select_best_score(grid):
    
    results = pd.DataFrame(grid.cv_results_)
    
    results.sort_values(by='mean_test_score', ascending=False, inplace=True)
    
    results = results[['mean_test_score', 'std_test_score']]
    
    return results.head(1)

#Fonksiyonu test edelim
select_best_score(rf_lineer)
```

```python
#Tüm modellerin en iyi skorlarından dataframe oluşturalım
results = pd.concat([ 
    select_best_score(elastic_lineer),
    select_best_score(elastic_monoton),
    select_best_score(elastic_agac),
    
    select_best_score(knn_lineer),
    select_best_score(knn_monoton),
    select_best_score(knn_agac),
    
    select_best_score(svm_lineer),
    select_best_score(svm_monoton),
    select_best_score(svm_agac),
    
    select_best_score(rf_lineer),
    select_best_score(rf_monoton),
    select_best_score(rf_agac),
    ], axis=0)

results.index = [
    'elastic_lineer','elastic_monoton', 'elastic_agac',
    'knn_lineer','knn_monoton', 'knn_agac',
    'svm_lineer','svm_monoton', 'svm_agac',
    'rf_lineer','rf_monoton', 'rf_agac',
]

results.sort_values(by='mean_test_score', ascending=False, inplace=True)
```

```python
#Sonuçlar
results
```

```python
results['mean_test_score'].plot.bar(
    yerr=[results['std_test_score'], results['std_test_score']],
    subplots=True, figsize=(10,5))

plt.ylabel('Skor')
plt.show()
```

En iyi skorun **elastic net** algoritmasıyla elde edilebildiğini görüyoruz. Ön işleme farklılıkları olan veri setleri arasında ise skorlar açısından herhangi bir fark yoktur. Böyle bir durumda hangi ön işleme yöntemi kullanarak hangi modeli seçeceğiz diye soracak olursak ve **basit her zaman iyidir** mottosundan yola çıkarsak:

1. Diğerlerine göre daha basit modelin seçilmesi ki bu örnekte zaten lineer model olan elastic net en iyi sonucu verdi.

2. Verileri işleme sürelerini göz önüne alırsak, değişken uzayını çok büyütmeyen ön işleme yöntemini seçebiliriz. Bu durumda lineer ön işlemede one-hot-encoding yapmamız nedeniyle değişken uzayı ön işleme yöntemlerine göre daha büyük.

3. Yine işlem yoğunluğu ve süresi ele alındığında, ağaç ön işlemede veri kesikleştirme ve kodlama işlemleri de bir model ile yapıldığı için, bu daha fazla işlem sayısı, dolayısıyla veri büyüdüükçe işlem yoğunluğu ve hızı bakımından olumsuz sonuçlarla karşılaşabiliriz.

**Sonuç:** Değişken uzayını büyütmeyen ve işlem yoğunluğu bakımından diğerlerine göre daha az yoğun olan monoton ön işleme ile birlikte elastic net algoritmasıyla modelimizi tamamlayabiliriz.
